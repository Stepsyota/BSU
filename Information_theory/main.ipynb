```python
df = df.drop(["Repository Name", "Full Name", "Description", "Owner Login", "Topics"], axis="columns")
df.info()
```

```python
for col in df.columns:
    number_unique = df[col].nunique()
    print(f"{col} : {number_unique}")
```

```python
num_cols = ['Forks Count', 'Watchers Count', 'Open Issues Count', 'Size (KB)']

for col in num_cols:
    print(f"{col}: min={df[col].min()}, max={df[col].max()}")
```

```python
# Преобразуем строки в datetime
df['Created At'] = pd.to_datetime(df['Created At']).dt.tz_localize(None)
df['Updated At'] = pd.to_datetime(df['Updated At']).dt.tz_localize(None)
df['Pushed At'] = pd.to_datetime(df['Pushed At']).dt.tz_localize(None)

# Минимум и максимум
date_cols = ['Created At', 'Updated At', 'Pushed At']
for col in date_cols:
    print(f"{col}: min={df[col].min()}, max={df[col].max()}")
```

```python
df['Forks_Bin'] = pd.qcut(df['Forks Count'], q=10, labels=False)
df['Watchers_Bin'] = pd.qcut(df['Watchers Count'], q=10, labels=False)
df['Open_Issues_Bin'] = pd.qcut(df['Open Issues Count'], q=10, labels=False)
df['Size(KB)_Bin'] = pd.qcut(df['Size (KB)'], q=10, labels=False)

today = pd.Timestamp.today()

# Переводим даты в количество дней от сегодня
df['Created_Days'] = (today - df['Created At']).dt.days
df['Updated_Days'] = (today - df['Updated At']).dt.days
df['Pushed_Days'] = (today - df['Pushed At']).dt.days

# Категоризация по бинам
df['Created_Bin'] = pd.qcut(df['Created_Days'], q=10, labels=False)
df['Updated_Bin'] = pd.qcut(df['Updated_Days'], q=6, labels=False)
df['Pushed_Bin'] = pd.qcut(df['Pushed_Days'], q=10, labels=False)
```

```python
import numpy as np

df = df.drop(['Forks Count', 'Watchers Count', 'Open Issues Count', 'Size (KB)', 'Created_Days', 'Updated_Days', 'Pushed_Days', 'Created At', 'Updated At', 'Pushed At'], axis="columns")

df['Stars Count'] = np.log1p(df['Stars Count'])
for col in df.columns:
    number_unique = df[col].nunique()
    print(f"{col} : {number_unique}")
```
