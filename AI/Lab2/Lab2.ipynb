```python
import pandas as pd

df = pd.read_csv("../Lab1/after_lab1.csv")
df = df.drop(["Name", "Ticket"], axis = "columns")
df.head(10)
```

```python
print(len(df))
```

```python
from sklearn.model_selection import train_test_split

X_reg = df.drop(["Age"], axis = "columns")
y_reg = df["Age"]
X_class = df.drop(["Survived"], axis = "columns")
y_class = df["Survived"]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size = 0.4, random_state = 42)
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_class, y_class, test_size = 0.4, random_state = 42)
```

```python
print(f"Len_X_train_reg = {len(X_train_reg)},Len_X_test_reg = {len(X_test_reg)}, Summa = {len(X_train_reg) + len(X_test_reg)} ")
print(f"Len_X_train_class = {len(X_train_class)},Len_X_test_class = {len(X_test_class)}, Summa = {len(X_train_class) + len(X_test_class)} ")
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train_reg, y_train_reg)
y_pred_test_reg = linear_model.predict(X_test_reg)

from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error
MSE = mean_squared_error(y_test_reg, y_pred_test_reg)
RMSE = root_mean_squared_error(y_test_reg, y_pred_test_reg)
MAE = mean_absolute_error(y_test_reg, y_pred_test_reg)
print(f"MSE = {MSE}, RMSE = {RMSE}, MAE = {MAE}")
```

```python
from sklearn.linear_model import LogisticRegression

logreg_model = LogisticRegression(max_iter = 1000)
logreg_model.fit(X_train_class, y_train_class)
y_pred_test_class = logreg_model.predict(X_test_class)

from sklearn.metrics import accuracy_score
accuracy_class = accuracy_score(y_test_class, y_pred_test_class)
print(accuracy_class)
```
