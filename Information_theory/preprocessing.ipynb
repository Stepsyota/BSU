```python
import pandas as pd

df = pd.read_csv("github_top_repositories.csv")

print(f"Rows before deduplication: {len(df)}")
print(f"Duplicate repositories by Full Name: {df['Full Name'].duplicated().sum()}")

df = df.drop_duplicates(subset='Full Name').copy()

print(f"Rows after deduplication: {len(df)}")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df = df.drop(["Repository Name", "Full Name", "Description", "Owner Login", "Topics"], axis="columns")
df.drop(['Watchers Count'], axis='columns', inplace=True) # Пришлось удалить из-за проблемы датасета, что Stars count == Watchers count
df.info()
```

```python
nan_matrix = df.isnull()
print(nan_matrix.sum())
```

```python
df.fillna({'Primary Language' : 'Unknown', 'License' : 'Unknown'},inplace=True)

nan_matrix = df.isnull()
print(nan_matrix.sum())
```

```python
for col in df.columns:
    number_unique = df[col].nunique()
    print(f"{col} : {number_unique}")
```

```python
num_cols = ['Forks Count', 'Open Issues Count', 'Size (KB)']

for col in num_cols:
    print(f"{col}: min={df[col].min()}, max={df[col].max()}")


df['Created At'] = pd.to_datetime(df['Created At']).dt.tz_localize(None)
df['Updated At'] = pd.to_datetime(df['Updated At']).dt.tz_localize(None)
df['Pushed At'] = pd.to_datetime(df['Pushed At']).dt.tz_localize(None)

date_cols = ['Created At', 'Updated At', 'Pushed At']
for col in date_cols:
    print(f"{col}: min={df[col].min()}, max={df[col].max()}")
```

```python
df['Forks_Bin'] = pd.qcut(df['Forks Count'], q=10, labels=False)
df['Open_Issues_Bin'] = pd.qcut(df['Open Issues Count'], q=10, labels=False)
df['Size(KB)_Bin'] = pd.qcut(df['Size (KB)'], q=10, labels=False)

reference_day = pd.Timestamp('2026-03-14')

# Переводим даты в количество дней от референс дня
df['Created_Days'] = (reference_day - df['Created At']).dt.days
df['Updated_Days'] = (reference_day - df['Updated At']).dt.days
df['Pushed_Days'] = (reference_day - df['Pushed At']).dt.days

# Категоризация по бинам
df['Created_Bin'] = pd.qcut(df['Created_Days'], q=10, labels=False)
df['Updated_Bin'] = pd.qcut(df['Updated_Days'], q=10, labels=False, duplicates='drop')
df['Pushed_Bin'] = pd.qcut(df['Pushed_Days'], q=10, labels=False, duplicates='drop')

df.drop(['Forks Count', 'Open Issues Count', 'Size (KB)', 'Created_Days',
     'Updated_Days', 'Pushed_Days', 'Created At', 'Updated At', 'Pushed At'], axis="columns", inplace=True)

for col in df.columns:
    number_unique = df[col].nunique()
    print(f"{col} : {number_unique}")
```

```python
# Бинарные признаки переводим в 0/1 для модели
bool_cols = ['Has Wiki', 'Has Pages', 'Has Projects']
df[bool_cols] = df[bool_cols].astype(int)

# Для ветки достаточно оставить main/master/other
df['Default Branch'] = df['Default Branch'].where(
    df['Default Branch'].isin(['main', 'master']),
    'other'
)

# Organization=1, User=0
df['Owner Type'] = (df['Owner Type'] == 'Organization').astype(int)

# One-hot encoding
categorical_cols = ['Domain', 'Primary Language', 'License', 'Default Branch']
df = pd.get_dummies(df, columns=categorical_cols, dtype=int, drop_first=True)

print(df.shape)
df.head()
```

```python
import numpy as np

df['Stars Count'] = np.log1p(df['Stars Count'])
print(df['Stars Count'])
```

```python
df.to_csv('processed_data.csv', index=False)
```
